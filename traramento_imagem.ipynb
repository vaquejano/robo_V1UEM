```python
#preparar as imagens 
imagem = (r"2.png")
img_cv = cv.imread(imagem)
plt.imshow(img_cv)
import numpy as np
```

```python
#passar para rgb
img = cv.cvtColor(img_cv, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img_gray = cv.cvtColor(img,cv.COLOR_RGB2GRAY)
plt.imshow(img_gray)
```

```python
plt.imshow(img_gray, cmap = plt.cm.gray)
```

```python
#TRANSLAÇÃO

altura, largura = img.shape[:2]
#deslocamento em x e y

deslocamento = np.float32([[1,0,-12],[0,1,-12]])
#cv.warpAffine
img_trans = cv.warpAffine(img, deslocamento,( largura, altura))
plt.imshow(img_trans)
```

```python
#rotação
ponto = (largura/2, altura/2) #pontocentral
rotacao = cv.getRotationMatrix2D(ponto, 180, 3.0)
img_rot = cv.warpAffine(img, rotacao,(largura, altura))
plt.imshow(img_rot)
```

```python
#espelhar imagem
espelhar = cv.flip(img,-1)
plt.imshow(espelhar)
```

```python
#FUNCAO PARA GERAR THRESHOLD AUTOMATICO
def auto_canny(mage,sigma=0.33):
     # compute the median of the single channel pixel intensities
     v = np.median(mage)

     # apply automatic Canny edge detection using the computed median
     lower = int(max(0, (1.0 - sigma) * v))
     upper = int(min(255, (1.0 + sigma) * v))
     edged = cv.Canny(mage, lower, upper)
     print(lower, upper)
     # return the edged image
     return edged
mg_canny = auto_canny(mg_gray, sigma=0.33)
plt.imshow(mg_canny)     
```
